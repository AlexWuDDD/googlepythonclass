```python
def halvesString(s):
    halfnum = len(s)//2
    if len(s)%2 == 0:
        return s[: halfnum], s[halfnum:]
    else:
        return s[: halfnum+1], s[halfnum+1:]
```

```python
halvesString('alex')
```

```python
def print_words(filename):
    worddict = {}
    with open(filename) as f:
        for word in f.read().split():
            if word in worddict:
                worddict[word.lower()] +=1
            else:
                worddict[word.lower()] = 0
    return worddict
print_words('alice.txt')
```

```python

def print_top(filename):
    worddict = print_words(filename)
    worddictSorted = sorted(worddict.items(), key=lambda item:item[1], reverse=True)
    return worddictSorted[:20]

print_top('Alice.txt')
```

```python
def mimic_dict(filename):
    mimic_dict = {}
    f = open(filename, 'r')
    text = f.read()
    f.close()
    words = text.split()
    #print(words)
    prev = ''
    for word in words:
        if not prev in mimic_dict:
            mimic_dict[prev] = [word]
        else:
            mimic_dict[prev].append(word)
        prev = word
    return mimic_dict
print(mimic_dict('small.txt'))
```

```python
import random
def print_mimic(mimic_dict, word):
    for unused_i in range(200):
        if word != 'We':
            print(word + ' ',end='') #打印输入的word
        else:
            print('')
            print(word + ' ', end='')
        nexts = mimic_dict.get(word)          # 获取key word 对应的value
        if not nexts:
            nexts = mimic_dict['']  # Fallback to '' if not found
        word = random.choice(nexts) #在 value里随便选一个当做word
dict = mimic_dict('small.txt')
print_mimic(dict, '')
```
